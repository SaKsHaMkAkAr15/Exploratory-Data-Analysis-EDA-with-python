```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.xlabel("0 = Died, 1 = Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival by Gender")
plt.xlabel("0=Died, 1=Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.title("Survival byPassenger Classr")
plt.xlabel( " 0=Died,   1=Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby("Survived")["Age"].mean().plot(kind="bar", color=["red", "green"])
plt.title("Average Age: Survived vs Died")
plt.xlabel("0 = Died, 1 = Survived")
plt.ylabel("Average Age")
plt.xticks(rotation=0)
plt.show()
```

```python
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title('Fare vs Survival')
plt.xlabel("0 = Died, 1 = Survived")
plt.ylabel("Fare Paid")
plt.show()
```

```python
print("<--- DATA ANALYSIS INSIGHTS OF TITANIC --->")
print(f"Total Passengers: {len(df)}")
print(f"Survival Rate: {round(df['Survived'].mean()*100, 2)}%")
print(f"Average Age: {round(df['Age'].mean(), 1)} years")
print(f"Average Fare: £{round(df['Fare'].mean(), 2)}")
print(f"Female Survival Rate: {round(df[df['Sex']=='female']['Survived'].mean()*100, 2)}%")
print(f"Male Survival Rate: {round(df[df['Sex']=='male']['Survived'].mean()*100, 2)}%")
print(f"1st Class Survival Rate: {round(df[df['Pclass']==1]['Survived'].mean()*100, 2)}%")
print(f"3rd Class Survival Rate: {round(df[df['Pclass']==3]['Survived'].mean()*100, 2)}%")
```
